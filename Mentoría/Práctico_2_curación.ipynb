5) Despersonalización de Datos.

---


Vamos a suponer que los id de los camiones que se presetan en el datast son datos que permiten identificar a los mismos y por un tema de confidencialidad necestiamos despersonalizar este dato para hacer público su análisis. Aplicar dos técnicas de despersonalización de datos sobre el dataset



```python
#Datos sin anonimizar
display(dataset2[['Id']].head(5))
```

```python
#Creamos una funcion para anonimizar
def anoni1 (column):
    return ((column/3333)-np.mod(13,7))
```

```python
#Anonimizamos los datos
dataset2[['Id']].apply(anoni1).head(5)
```

```python
#Creamos otra funcion para anonimizar
def anoni2 (column):
    return ((column/2222)-np.mod(14,6))
```

```python
dataset2[['Id']].apply(anoni2).head(5)
```

```python
#Le aplicamos la anonimidad a todo el data set
dataset2[['Id']] = dataset2[['Id']].apply(anoni1)
```

*Vale aclarar que existen diferentes (infinitas) formas de "Anonimizar" los datos, ya que a fin de cuenta lo que se aplica es una 'función' que transforme los mismos. Ej: MD5,SHAx,HILL...*

7)Valores faltantes

Determinar que porcentaje de datos faltantes existen en el feature Fe y Na
Completar los datos faltantes con la media

```python
dataset[['Id','Fe','Na']].count()
```

```python
dataset[dataset.duplicated(['Id'])]
```

```python
total_datos = dataset[['Id']].count().values[0]+1
total_Fe = dataset[['Fe']].dropna().count().values[0]
total_Na = dataset[['Na']].dropna().count().values[0]

porc_falt_Fe = abs(((100*total_Fe)/total_datos)-100)
porc_falt_Na = abs(((100*total_Na)/total_datos)-100)
display('Total de datos',total_datos, 'Total de datos de Hiero', total_Fe, 'Porcentaje faltante de datos de Hierro (Fe)', porc_falt_Fe, 'Total de datos de Sodio',total_Na, 'Porcentaje faltante de datos de Sodio (Na)',porc_falt_Na)
```

```python
#Completamos los datos faltantes de hierro y Sodio con sus respectivas medias
display(dataset[['Na']].fillna(dataset[['Na']].mean()).dropna().count(),dataset[['Fe']].fillna(dataset[['Fe']].mean()).dropna().count())
dataset2[['Na']] = dataset[['Na']].fillna(dataset[['Na']].mean()).dropna()
dataset2[['Fe']] = dataset[['Fe']].fillna(dataset[['Fe']].mean()).dropna()
```
